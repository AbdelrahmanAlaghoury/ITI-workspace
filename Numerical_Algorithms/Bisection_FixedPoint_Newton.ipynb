### Bisection Method

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*x**2 + 6*x + 4

def bisection(f, istart, iend, epislon):
    g_list = []
    while iend - istart > epislon:
        g = (istart + iend) / 2 
        if f(g) * f(iend) <= 0:
            istart = g
        else:
            iend = g
        
        g_list.append(g)
    return g_list



        
g_list = bisection(f, -3, -1.5, 1e-5)    
        
        
x = np.linspace(-3, 0)
plt.axhline(y=0)
plt.plot(x, f(x),'r')
plt.show()

plt.axhline(y=0)
for g in g_list:
    plt.axvline(x = g)
plt.plot(x, f(x),'r')
plt.show()
print("f(g) = ", g_list[-1])
```

### Fixed Point Method

```python
def g(x):
    return (-2*x**2 - 4)/ 6

def fixedPoint(g, startPoint, epsilon):
    all_points = []
    p1 = g(startPoint)
    while(abs(p1 - startPoint) > epsilon):
        all_points.append(startPoint)        
        startPoint = p1
        p1 = g(startPoint)
    return all_points

all_points = fixedPoint(g, -1.8, 1e-5)
x = np.linspace(-3,0)
plt.plot(x, g(x))
plt.plot(x, x)
for point in all_points:
    plt.scatter(point, g(point))
plt.show()
print(all_points[-1])
```

### Newton method

```python
def f(x):
    return 2*x**2 + 6*x + 4

def df(x):
    return 4*x + 6

def newtonMethod(f, df, x0, epsilon):
    all_x = []
    x1 = x0 - f(x0)/df(x0)
    while abs(x1 - x0) > epsilon:
        all_x.append(x0)
        x0 = x1
        x1 = x0 - f(x0)/df(x0)
    return all_x

all_x = newtonMethod(f, df, -3.0, 1e-8)

x = np.linspace(-3, 0)
plt.plot(x, f(x))
for point in all_x:
    plt.scatter(point, f(point))
plt.show()
    
print(all_x)
```
