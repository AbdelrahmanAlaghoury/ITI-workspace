```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

## Quadratic function

```python
def Q(x):
    x1, x2 = x
    return 1.5*x1**2 + x1*x2 + x2**2 + x1 + x2
Q([0,0])
```

```python
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x1, x2)
Z = Q([X,Y])

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X, Y, Z)
# ax2 = fig.add_subplot(111, projection = '3d')
# ax2.contour3D(X, Y, Z, levels = 20)

plt.show()

plt.figure()
plt.contour(X, Y, Z, levels=20)
plt.show()
```

## Powell method

```python
from scipy.optimize import minimize

res = minimize(Q, np.array([-4, 4]), method="Powell", options={'return_all':True})
all_points = np.array(res.allvecs)

plt.figure()
plt.contour(X, Y, Z, levels=20)
plt.scatter(all_points[:,:1], all_points[:,1:], c='r', marker='x')
plt.plot(all_points[:,:1], all_points[:,1:])
plt.show()


# print(res)
```
